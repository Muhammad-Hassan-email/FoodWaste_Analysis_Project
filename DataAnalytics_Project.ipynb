```python
pk_rows = data[data['Country'] == 'Pakistan']   # stays a DataFrame
```

```python
house_hold_tons = pk_rows['Household estimate (tonnes/year)']
```

```python
house_hold_tons
```

```python
retail_est = pk_rows['Retail estimate (tonnes/year)']
```

```python
retail_est
```

```python
food_est = pk_rows['Food service estimate (tonnes/year)']
```

```python
food_est
```

```python
total_ton = house_hold_tons + retail_est + food_est
```

```python
total_ton
```

```python
truck_capacity = 10  # Average capacity in tonnes
trucks_needed = total_ton / truck_capacity
```

```python
print(f"House Hold: {house_hold_tons}")
print(f"Retail Estimated Waste: {retail_est}")
print(f"Food Service Waste: {food_est}")
#print(f"Trunk Needed: {int(trucks_needed)}")
print(f"Trucks Needed: {int(trucks_needed.iloc[0]):,}")
```
